## Beautiful Soup
Beautiful Soup is a Python library for pulling data out of HTML and XML files. It works with a parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. it transforms a complex HTML document into a complex tree of Python objects which are Tag, NavigableString, BeautifulSoup, and Comment.

Web scraping is the process of gathering information from the Internet. Even copying and pasting  is a form of web scraping! However, the words “web scraping” usually refer to a process that involves automation. Some websites don’t like it when automatic scrapers gather their data, while others don’t mind.

__find_all(name, attrs, recursive, text, limit, **kwargs)__

* **name**: The tag name or a list of tag names to be searched.
* **attrs**: A dictionary of attributes and their corresponding values to filter elements.
* **recursive**: A Boolean value to specify whether to search only the direct children or the entire descendants (default is True).
* **text**: A string or regular expression to find elements containing specific text.
* **limit**: An integer specifying the maximum number of elements to return.
* __**kwargs__: Allows the use of CSS selectors or other filters for specific use cases.

```python
# Create function to find a tag with no id.
def has_class_but_no_id(tag):
    return tag.has_attr('class') and not tag.has_attr('id')

print(soup.find_all(has_class_but_no_id))
```

```python
# Function that finds all a tags whose href attribute does not match a regular expression.
import re
def not_article(href):
    return href and not re.compile("article").search(href)

soup.find_all(href=not_article)
```

```python
# Function that finds tags with string.
from bs4 import NavigableString
def surrounded_by_strings(tag):
    return (isinstance(tag.next_element, NavigableString)
            and isinstance(tag.previous_element, NavigableString))

for tag in soup.find_all(surrounded_by_strings):
    print(tag.name)
```

```python
# Find all tags with arguments.
match = soup.find_all(id=True)
print(match)
```

```python
# Some attributes can be used as the names of keyword arguments.
match = soup.find_all(data="end")
print(match)
```

```python
# Alternatively, put them into a dictionary and pass it as the attrs argument.
match = soup.find_all(attrs={"data": "end"})
print(match)
```

#### After the procedure above, a while loop will be created to extract all 20 pages of the website.

```python
# Create a program to scrap the whole website.
from bs4 import BeautifulSoup
import requests

page = 1

title_list = []
type_list = []
summary_list = []
author_list = []
date_list = []
read_list = []


while page < 79:
    
    website = f'https://stackabuse.com/tag/python/page/{page}/'
    source = requests.get(website).text
    soup = BeautifulSoup(source, 'lxml')
        
    
    for data in soup.find_all('div', class_="flex-1 bg-white p-6 flex flex-col justify-between"):
        
        # Get the title.
        title_find = data.find('h3')
        title_string = title_find.string
        title_list.append(title_string)


        # Get the type.
        type_find = data.find('p', class_="text-sm leading-5 font-medium text-primary")
        type_string = type_find.string
        type_list.append(type_string)


        # Get the summary.
        summary_find = data.find('p', class_="mt-3 text-base leading-6 text-gray-500 break-words")
        summary_string = summary_find.string
        summary_list.append(summary_string)


        # Get the author.
        author_find = data.find('a', class_="hover:underline")
        author_string = author_find.string
        author_list.append(author_string)


        # Get the date.
        date_find = data.find('time')
        date_string = date_find.string
        date_list.append(date_string)


        # Get the read.
        read_find = data.find('span', class_="mx-1")

        if read_find == None:
            read_list.append('NONE')

        else:
            read_sibling = read_find.find_next_sibling()
            read_string = read_sibling.text
            read_list.append(read_string)

    
    print(f'Scraped page -----> {page}')
    page += 1
```

```python
# Check the length of the extracted items; they must be equal.
items_len =[f'title_list: {len(title_list)}', 
            f'type_list: {len(type_list)}', 
            f'summary_list: {len(summary_list)}',
            f'author_list: {len(author_list)}',
            f'date_list: {len(date_list)}',
            f'read_list: {len(read_list)}']

for item in items_len:
    print(item)
```

```python
# Create a dictionary of the extracted items.
scrap_dict = {'title': title_list,
              'type': type_list,
              'summary': summary_list,
              'author': author_list,
              'date': date_list,
              'read': read_list}
```

```python
# Create a dataframe from the dictionary.
import pandas as pd
scrap_df = pd.DataFrame(scrap_dict)
```
